```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('../src/benin-malanville.csv')
data.head()
```

```python
cols = data.columns
count = 0
for i in cols:
    column_dtype = data[i].dtype

    # Check if the column has a numeric dtype
    is_numeric = column_dtype in ['int64', 'float64']
    count += 1

    print("Is numeric:", is_numeric)
print(f"Number of cols:{count}")
```

```python
cols = data.columns
count = 0
for i in cols:
    is_unique = data[i].unique()
    if len(is_unique) < 20:
        count += 1

        print(f"col:{i} how many unique:", is_unique)
print(f"Number of cols:{count}")
```

```python
for i in data.columns:
    if i != "Timestamp":
        # Calculate mean
        mean_value = data[i].mean()

        # Calculate median
        median_value = data[i].median()

        # Calculate standard deviation
        std_deviation = data[i].std()

        # Calculate minimum
        min_value = data[i].min()

        # Calculate maximum
        max_value = data[i].max()

        # Calculate count
        count = data[i].count()
        
        print("col name:", i)
        print("Mean:", mean_value)
        print("Median:", median_value)
        print("Standard Deviation:", std_deviation)
        print("Minimum:", min_value)
        print("Maximum:", max_value)
        print("Count:", count)
        print(" ")
    else:
        continue 
```
