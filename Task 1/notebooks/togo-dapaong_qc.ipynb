```python
from scipy import stats

# Calculate Z-score
z_scores = stats.zscore(cleaned_data[['GHI', 'DNI', 'DHI']])
print(z_scores)

# Identify outliers based on Z-score
outliers = (np.abs(z_scores) > 3).any(axis=1)

# Print rows with outliers
print(cleaned_data[outliers])

# Drop outliers from the original DataFrame
# Drop outliers from the original DataFrame
cleaned_data = cleaned_data.drop(cleaned_data[outliers].index)

# Print the cleaned DataFrame
print(cleaned_data)

```

```python
import matplotlib.pyplot as plt

# Convert 'Timestamp' column to datetime type
cleaned_data['Timestamp'] = pd.to_datetime(cleaned_data['Timestamp'])

# Set 'Timestamp' column as the index
cleaned_data.set_index('Timestamp', inplace=True)

# Plot the variables over time
plt.figure(figsize=(12, 8))
plt.plot(cleaned_data['GHI'], label='GHI')
plt.plot(cleaned_data['DNI'], label='DNI')
plt.plot(cleaned_data['DHI'], label='DHI')
plt.plot(cleaned_data['Tamb'], label='Tamb')
plt.xlabel('Timestamp')
plt.ylabel('Value')
plt.title('Time Series Analysis (Cleaned Data)')
plt.legend()
plt.show()
```

```python
# Calculate correlation matrix
correlation_matrix = cleaned_data[['GHI', 'DHI', 'DNI', 'TModA', 'TModB']].corr()

# Print correlation matrix
print(correlation_matrix)
```

```python
import matplotlib.pyplot as plt

# Visualize Wind Speed Trends
plt.figure(figsize=(12, 6))
plt.plot(cleaned_data.index, cleaned_data['WS'], label='WS')
plt.plot(cleaned_data.index, cleaned_data['WSgust'], label='WSgust')
plt.plot(cleaned_data.index, cleaned_data['WSstdev'], label='WSstdev')
plt.xlabel('Timestamp')
plt.ylabel('Wind Speed (m/s)')
plt.title('Wind Speed Trends')
plt.legend()
plt.show()

# Visualize Wind Direction Trends
plt.figure(figsize=(12, 6))
plt.plot(cleaned_data.index, cleaned_data['WD'], label='WD')
plt.plot(cleaned_data.index, cleaned_data['WDstdev'], label='WDstdev')
plt.xlabel('Timestamp')
plt.ylabel('Wind Direction (degrees)')
plt.title('Wind Direction Trends')
plt.legend()
plt.show()

# Wind Speed Statistics
wind_speed_stats = cleaned_data[['WS', 'WSgust', 'WSstdev']].describe()
print("Wind Speed Statistics:")
print(wind_speed_stats)

# Wind Direction Statistics
wind_direction_stats = cleaned_data[['WD', 'WDstdev']].describe()
print("\nWind Direction Statistics:")
print(wind_direction_stats)

# Correlation Analysis
wind_correlation = cleaned_data[['WS', 'WSgust', 'WSstdev', 'WD', 'WDstdev']].corr()
print("\nCorrelation Analysis:")
print(wind_correlation)
```

```python
import matplotlib.pyplot as plt

# Visualize Temperature Trends
plt.figure(figsize=(12, 6))
plt.plot(cleaned_data.index, cleaned_data['TModA'], label='TModA')
plt.plot(cleaned_data.index, cleaned_data['TModB'], label='TModB')
plt.plot(cleaned_data.index, cleaned_data['Tamb'], label='Tamb')
plt.xlabel('Timestamp')
plt.ylabel('Temperature (°C)')
plt.title('Temperature Trends')
plt.legend()
plt.show()

# Scatter Plot for TModA vs. Tamb
plt.figure(figsize=(8, 6))
plt.scatter(cleaned_data['Tamb'], cleaned_data['TModA'], alpha=0.5)
plt.xlabel('Ambient Temperature (°C)')
plt.ylabel('Module Temperature (TModA) (°C)')
plt.title('Scatter Plot: TModA vs. Tamb')
plt.grid(True)
plt.show()

# Scatter Plot for TModB vs. Tamb
plt.figure(figsize=(8, 6))
plt.scatter(cleaned_data['Tamb'], cleaned_data['TModB'], alpha=0.5)
plt.xlabel('Ambient Temperature (°C)')
plt.ylabel('Module Temperature (TModB) (°C)')
plt.title('Scatter Plot: TModB vs. Tamb')
plt.grid(True)
plt.show()

# Correlation Analysis
temperature_correlation = cleaned_data[['TModA', 'TModB', 'Tamb']].corr()
print("\nCorrelation Analysis:")
print(temperature_correlation)
```
